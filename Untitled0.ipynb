```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Simple dataset (house size vs price)
data = {
    "size": [500, 800, 1000, 1200, 1500, 1800, 2000],
    "price": [30, 50, 65, 80, 100, 120, 140]
}

df = pd.DataFrame(data)

# 2. Split data
X = df[["size"]]
y = df["price"]

# 3. Create model
model = LinearRegression()
model.fit(X, y)

# 4. Predict
df["predicted_price"] = model.predict(X)

# 5. Print result
print(df)

# 6. GRAPH (IMPORTANT PART)
plt.scatter(df["size"], df["price"], color="blue", label="Actual Price")
plt.plot(df["size"], df["predicted_price"], color="red", label="Prediction Line")

plt.title("House Price Prediction")
plt.xlabel("House Size")
plt.ylabel("Price")
plt.legend()

plt.show()
```

**This project is a simple machine learning model that predicts house prices based on house size.**

It uses a method called linear regression, which finds a straight line that best fits the relationship between size and price. The idea is that larger houses generally have higher prices, and smaller houses have lower prices.

In the program, we first create a small dataset with house sizes and their corresponding prices. Then the model learns from this data and tries to understand the pattern between the two values. After training, it is able to predict prices for the same house sizes.

Finally, a graph is shown where blue points represent the actual house prices, and the red line represents the predicted trend created by the model. This helps us visually understand how well the model has learned the relationship between size and price.

Overall, the project demonstrates the basic idea of how machine learning can be used to make predictions from data.

```python
import matplotlib.pyplot as plt

# past years
years = [2019, 2020, 2021, 2022, 2023]

# actual house prices
prices = [60, 65, 70, 85, 100]

# predicted future years
future_years = [2024, 2025, 2026]

# predicted prices (model trend continued)
future_prices = [115, 130, 145]

# plot past data
plt.plot(years, prices, marker="o", label="Actual Prices")

# plot prediction
plt.plot(future_years, future_prices, marker="o", linestyle="dashed", color="red", label="Predicted Prices")

plt.title("House Price Prediction Over Time")
plt.xlabel("Year")
plt.ylabel("Price (in $1000s)")
plt.legend()

plt.show()
```

This prediction helps estimate future market growth based on past patterns.

**If house prices keep increasing, how much extra money will I lose by waiting 3 years?**

```python
import matplotlib.pyplot as plt

# past house prices (in $1000s)
years = [2019, 2020, 2021, 2022, 2023]
prices = [60, 65, 70, 85, 100]

# simple yearly growth
growth = (prices[-1] - prices[0]) / (len(prices) - 1)

# future prediction
future_years = [2024, 2025, 2026]
future_prices = []
last = prices[-1]

for i in range(3):
    last += growth
    future_prices.append(last)

# FULL DATA COMBINE
all_years = years + future_years
all_prices = prices + future_prices

# -------------------------
# GRAPH 1: FULL TREND
# -------------------------
plt.figure()
plt.plot(all_years, all_prices, marker="o")
plt.title("House Price Trend (Past + Future)")
plt.xlabel("Year")
plt.ylabel("Price ($1000s)")
plt.show()

# -------------------------
# GRAPH 2: MONEY LOSS IF YOU WAIT
# -------------------------
wait_years = [1, 2, 3]
extra_cost = []

for i in range(3):
    extra_cost.append(future_prices[i] - prices[-1])

plt.figure()
plt.bar(wait_years, extra_cost)
plt.title("Extra Cost If You Delay Buying")
plt.xlabel("Years Delay")
plt.ylabel("Extra Money ($1000s)")
plt.show()

# -------------------------
# GRAPH 3: COMPARISON (NOW vs FUTURE)
# -------------------------
plt.figure()
plt.plot(future_years, future_prices, marker="o", color="red", label="Future Price")
plt.axhline(y=prices[-1], color="blue", linestyle="--", label="Current Price")

plt.title("Buy Now vs Wait Comparison")
plt.xlabel("Year")
plt.ylabel("Price ($1000s)")
plt.legend()
plt.show()
```

This project analyzes house price growth over time and solves a real decision problem: whether to buy now or wait.

The first graph shows historical and predicted price trends.
The second graph shows the extra cost if the buyer delays purchasing a house.
The third graph compares current price with future prices to help make a financial decision.

The analysis shows that delaying purchase leads to higher costs due to increasing market prices.
